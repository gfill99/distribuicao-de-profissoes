```python
import duckdb as db
import pandas as pd
```

```python
# Conectar ao banco de dados
con = db.connect("../database/db_distribuicao_profissoes.db")
```

```python
# Função para carregar as tabelas
def carregar_tabela(nome_tabela):
    return con.execute(f"SELECT * FROM {nome_tabela}").df()
```

```python
# Carregar as tabelas
origens_advogados = carregar_tabela("origens_advogados")
origens_contadores = carregar_tabela("origens_contadores")
origens_engenheiros = carregar_tabela("origens_engenheiros")
origens_psicologos = carregar_tabela("origens_psicologos")
```

```python
# Exibir as primeiras linhas de cada tabela
display(origens_advogados.head())
display(origens_contadores.head())
display(origens_engenheiros.head())
display(origens_psicologos.head())
```

```python
# Lista dos arquivos
arquivos = ["tbl_advogados_brasil_2024.xlsx", 
            "tbl_contadores_brasil_2024.xlsx", 
            "tbl_engenheiros_brasil_2024.xlsx", 
            "tbl_psicologos_brasil_2024.xlsx"
]

# Carregar os arquivos do Excel em um dicionário de DataFrames
tabelas_excel = {nome: pd.read_excel(f"../origens/{nome}") for nome in arquivos}
```

```python
# Acessar as tabelas carregadas
df_advogados = tabelas_excel["tbl_advogados_brasil_2024.xlsx"]
df_contadores = tabelas_excel["tbl_contadores_brasil_2024.xlsx"]
df_engenheiros = tabelas_excel["tbl_engenheiros_brasil_2024.xlsx"]
df_psicologos = tabelas_excel["tbl_psicologos_brasil_2024.xlsx"]
```

```python
# Exibir as primeiras linhas para validação
display(df_advogados.head())
display(df_contadores.head())
display(df_engenheiros.head())
display(df_psicologos.head())
```

```python
def validar_ingestao(df_tabela, df_arquivo):
    colunas_iguais = len(df_tabela.columns) == len(df_arquivo.columns)
    linhas_iguais = len(df_tabela) == len(df_arquivo)
    
    if colunas_iguais and linhas_iguais:
        print("Validação bem-sucedida! Os DataFrames são equivalentes em colunas e linhas.")
    else:
        print("Validação falhou.")
        if not colunas_iguais:
            print(f"Colunas diferentes: {len(df_tabela.columns)} (tabela) vs {len(df_arquivo.columns)} (arquivo)")
        if not linhas_iguais:
            print(f"Linhas diferentes: {len(df_tabela)} (tabela) vs {len(df_arquivo)} (arquivo)")
```

```python
validar_ingestao(origens_advogados, df_advogados)
validar_ingestao(origens_contadores, df_contadores)
validar_ingestao(origens_engenheiros, df_engenheiros)
validar_ingestao(origens_psicologos, df_psicologos)
```
